# An Example workflow to visualize module in a large network

```python
# Load libraries
from IPython.html.widgets import interact
from IPython.html.widgets import *
from IPython.html import widgets
import cytoscape.viewer as cy
import json
import networkx as nx
from igraph import *

import requests
import pandas as pd

# Make sure Cytoscape  RESTful API App is running!
cytoscape_version = requests.get('http://localhost:8080/v1/version')
print("Cytoscape RESTful API module is ready: version " +  json.loads(cytoscape_version.text)["cytoscapeVersion"])
```

## Import (relatively) large network as DataFrame

```python
yeast_df = pd.read_csv("data/BIOGRID-ORGANISM-Saccharomyces_cerevisiae-3.2.114.mitab.txt", delimiter="\t")
```

## Reformat and Convert into igraph object

```python
extract = lambda x: x.split("|")[0].split(":")[1]
yeast_df["source"] = yeast_df["Alt IDs Interactor A"].apply(extract)
yeast_df["target"] = yeast_df["Alt IDs Interactor B"].apply(extract)

yeast_df.head()
```

```python
to_tuple = lambda x, y: (x, y)

yeastnet = Graph()
nodes1 = set(yeast_df["source"].tolist())
nodes = nodes1.union(set(yeast_df["target"].tolist()))

node_list = list(nodes)

yeastnet.add_vertices(node_list)
yeastnet.add_edges(list(map(to_tuple, yeast_df["source"], yeast_df["target"] )))
yeastnet.get_edgelist()
```

```python
# Remove all networks
requests.delete('http://localhost:8080/v1/networks')

headers = {'Content-Type': 'application/json'}

small = {
    "data": {
        "name": "foo"
    },
    "elements":{
        "nodes":[],"edges":[]
    }
}

cyjs_network = cy.from_igraph(yeastnet)
res1 = requests.post("http://localhost:8080/v1/networks?collection=Yeast2", data=json.dumps(small), headers=headers)
```
